```python
import xml.etree.ElementTree
e = xml.etree.ElementTree.parse('A-shadow/104.xml').getroot()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

%matplotlib inline

for image in e:
    for words in image:
        print(words.tag, words.attrib)
        if words.tag == 'words':
            for word in words:
                print(word.tag, word.attrib)
                for char in word:
                    print(char.tag, char.attrib)
                    box = (int(char.attrib['x']), int(char.attrib['y']), int(char.attrib['x']) + int(char.attrib['width']), int(char.attrib['y']) + int(char.attrib['height']))
                    print(box)
                    region = im.crop(box)
#                     region.show()
                    plt.figure()
                    plt.imshow(np.asarray(region))
                    plt.show()
```
